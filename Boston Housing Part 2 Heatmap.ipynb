```python
import pandas as pd
import numpy as np
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
data = pd.DataFrame(boston.data)
```

```python
data
```

```python
data.columns = boston.feature_names
```

```python
data['PRICE'] = boston.target 
```

```python
corr = data.corr() #.corr() is used to find the pairwise correlation of all columns in the dataframe
corr.shape #finding out the shape the the table of correlation
#this is required string for the heatmap
```

```python
sns.heatmap(data,cbar=True, cmap="YlGnBu")

# data = retangular dataset
# 2D dataset that can be coerced into an ndarray. If a Pandas DataFrame is provided, 
# the index/column information will be used to label the columns and rows.
```

```python
sns.heatmap(corr, cmap="YlGnBu") #sns.heatmap uses seaborns function "heatmap"
```

```python
#plotting the heatmap of correlation between features
plt.figure(figsize=(20,20))
sns.heatmap(corr,
            linewidths=0, 
            cbar=True, 
            square=True, 
            fmt='.1f',
            annot=True, 
            annot_kws={'size':15}, 
            cmap='Reds')
           
#the 'corr' is used , as the correlation of the data set is assigned to the corr variable
#they can all be on the same line, but are on different lines = more readable
```

```python
#Functions of strongs that affect sns.heatmap
#cmap = colour map, they are a feature of Matplotlib eg cmap=Blues or cmap=bwr
#cbar = colour bar, whether to draw a colourbar or not
#cbar_kws = arguements for when cbar is true
#linewidths = width of the lins the divide each cell, float
#square = if True, set the Axes aspect to â€œequalâ€ so each cell will be square-shaped.
#annot = if true, write the data value in each cell
#annot_kws = arguements for when annot is true  (arguements modify behavior of a program)
```
